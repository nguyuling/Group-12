### Session 2: Exercise ###
*Consider the dna sequence, AGTTAGCTAGGAG*
* How many nucleotides are present?
* Calculate the percentage GC in the sequence.
* How many purines are present?
* How many pyrimidines are present?
* Calculate the percentage of purines in the sequence

```python
DNA = "AGTTAGCTAGGAG"
dna = DNA.lower()

# How many nucleotides are present?
count_necleotides = len(dna)

# Calculate the percentage GC in the sequence.
count_gc = dna.count('g') + dna.count('c')
percent_gc = count_gc / count_necleotides * 100

# How many purines are present?
count_purines = dna.count('a') + dna.count('g')

# How many pyrimidines are present?
count_pyrimidines = dna.count('c') + dna.count('t')

# Calculate the percentage of purines in the sequence
percent_purines = count_purines / count_necleotides * 100

print(f"No. of nucleotides present: {count_necleotides}")
print(f"Percentage of GC: {percent_gc} %")
print(f"No. of purines: {count_purines}")
print(f"No. of pyrimidines: {count_pyrimidines}")
print(f"Percenrage of purines: {percent_purines} %")
```

### Session 7: Exercise ###
*   Randomly generate a sequence of dna with length of 30.
*   Convert the DNA into RNA by replacing all occurrences of Thymine ('T') with Uracil ('U') while keeping other bases (Adenine 'A', Guanine 'G', Cytosine 'C') the same.
*   make a function to perform the convertion.


```python
from random import choice

# function to generate a random dna sequence
def generate_dna_sequence():
    length = 30
    bases = ['A', 'G', 'T', 'C']
    
    # randomly generate a dna sequence of given length
    dna = [choice(bases) for i in range(length)]
    
    # convert dna list to string and separate codons
    dna_str = "-".join(["".join(dna[i:i+3]) for i in range(0, length, 3)])
    return dna_str

# function to convert dna to rna
def convert_dna_to_rna(dna_str):
    rna_str = dna_str.replace('T', 'U')
    return rna_str

# function calling
dna_sequence = generate_dna_sequence()
rna_sequence = convert_dna_to_rna(dna_sequence)
print("DNA Sequence: ", dna_sequence)
print("RNA Sequence: ", rna_sequence)
```
