# Python Variable

Take variable as a container to store data values. Unlike other programming language, python doesn't has command to declare a variable or specify the data type. A variable is created at the moment we assign a value to it. `=` is the assign operator in python programming language.


```python

    x = 5
    y = "five"

```

In order to retrieve the value out from a variable, we can use `print` method. `print` method is different between `python2.7` and `python3.7`

For python 2.7

```python
    print x
    print y
```

For python 3.7

```python
    print(x)
    print(y)
```

## Retrieve Value from a List by Index

We are able to retrieve value from a `List` variable by `index`

### Example

```python
    # Index        0             1
    trainers = ["Dr.Sean", "Prof.Azlan"]
    
    print(trainers[0]
```
Try it yourself

````python
print(singers[0])
print(singers[1])
print(singers[2])
````

## Retrieve Index from a List by Value

By using `value`, we are able to get the index from a `List` variable


### Example

```python
    trainers = ["Dr.Sean", "Prof.Azlan"]
    print(trainers.index("Dr.Sean"))

```

````python
print(singers.index("Morgan Wallen"))
print(singers.index("Luke Comb"))
print(singers.index("Zack Bryan"))
````

## Tutorial: Rock Paper Scissors ( Step 1 )

In this tutorial, you are going to create a mini Python program that plays the classic Rock, Paper, Scissors game.

Let's follow the following steps:

Firstly, import a library, random
1. Create a variable named `options` and assign `rock`, `paper` and `scissors` to it

2. Create a variable named `user_counter` and assign 0 (zero) to it

3. Create a variable named `computer_counter` and random result to it


````python
import random

options = ["Rock","Paper","Scissors"]
user_choice = 0
computer_choice = random.choice(options)
````

## Tutorial: Rock Paper Scissors ( Step 3 )

Since we had created the variables, next try to rules the game, use **if else statement** from the following `code`.

````python
if user_choice == computer_choice:

    print("It's a tie!")

elif user_choice == "Rock" and computer_choice == "Scissors":

    print("You win!")

elif user_choice == "Paper" and computer_choice == "Rock":

    print("You win!")

elif user_choice == "Scissors" and computer_choice == "Paper":

    print("You win!")

else:

    print("Computer wins!")
````
