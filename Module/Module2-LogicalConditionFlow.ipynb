# Logical & Condition Flow

Truth Table, a mathematical table used in logic specifically in connection with Boolean algebra, boolean functions and propositional calculus which sets out the functional values of logical expressions on each of their functional arguments.

## Logical Conjunction (AND)

An operation on two logical values, typically the values of two propositions, that produces a value of true if both of its operands are ture.

The truth table for `p AND q`

| p | q | p and q|
|---|---|---|
| True | True | True |
| True | False | False |
| False| True | False |
| False| False | False |

Try printing out the result

````python
print(f"Result True and True: {True and True}")
print(f"Result True and False: {True and False}")
print(f"Result False and True: {False and True}")
print(f"Result False and False: {False and False}")
````

## Logical Disjuntion (OR)

An operation on two lofical values, typically the values of two propositions, that produces a value of true if at least one if its operands is true.

The truth table for `p OR q`

| p | q | p or q|
|---|---|---|
| True | True | True |
| True | False | True |
| False| True | True |
| False| False | False |

Task: Replicate the code above and print out the result

````python
print(f"Result True or True: {True or True}")
print(f"Result True or False: {True or False}")
print(f"Result False or True: {False or True}")
print(f"Result False or False: {False or False}")
````

## Logical Negation

An operation on one logical value, typically the value of a proposition, that produces a value of true if its operand is false and a value of false if its operand is true.

The truth table for `NOT p`

| p | not p |
|---|---|
| True | False |
| False | True |

Task: Replicate the code above and print out the result

````python
print(f"Result not True: {not True}")
print(f"Result not False: {not False}")
````

## Condition Flow

An `if` statement can be followed by an optional `else` statement, which executes when the Boolean expression is false.

### Example


**if** I don't know Python

    Proceed to intermediate level
    
### Python Code    
```python

if know_python:
    print("Learn Python")
else:
    print("Proceed to intermediate level")
```

Follow the Steps here:

3. Check the condition and determine who is winning
4. Add 1 to the counter ( `user_counter` or `computer_counter` )

Logic to determine who's winning

```python

    # index      0        1          2    
    options = ["rock", "paper", "scissors"]
    
    # player rock & computer scissors (player win) 0 - 2 = -2
    # player rock & computer paper (player lose) 0 - 1 = -1
    #
    # player paper & computer rock (player win) 1 - 0 = 1
    # player paper & computer scissors (player lose) 1 - 2 = -1
    #
    # player scissors & computer paper (player win) 2 - 1 = 1
    # player scissors & computer rock (lose) 2 - 0 = 2
    #
    # Player win when -2 and 1
    # Computer win when-1 and 2
```

````python
import random
options = ["rock", "paper", "scissors"]

computer_choice = random.choice(options)
computer_choice_index = options.index(computer_choice)

user_choice = input("Choose rock / paper / scissors: ")
user_choice_index = options.index(user_choice)

print (f"Computer choice: {computer_choice}")
print (f"User choice: {user_choice}")

outcome = user_choice_index - computer_choice_index

if outcome == -2 or outcome == 1:
    print("User win!")
elif outcome == 2 or outcome == -1:
    print("Computer win!")
elif outcome == 0:
    print("Draw!")
else:
    print("An error occurred!")
````
